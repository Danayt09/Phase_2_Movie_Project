## Second data: IMDB Info  

- (web link:https://www.imdb.com/)

- This is a database containing 8 tables. 
    

```python
import os
import zipfile

# Check if 'im.db' exists, unzip if it doesn't
if not os.path.exists('zippedData/im.db'):
    with zipfile.ZipFile('zippedData/im.db.zip', 'r') as zip_ref:
        zip_ref.extractall('zippedData')

# Connect to the SQLite database
conn = sqlite3.connect('zippedData/im.db')
```

```python
pd.read_sql("""
  SELECT *
  FROM sqlite_master  
  """, conn)
```

```python
movie_basics=pd.read_sql("""
  SELECT *
  FROM movie_basics
  """, conn)
movie_basics.info()
```

```python
directors=pd.read_sql("""
  SELECT *
  FROM directors
  """, conn)
directors.info()
```

```python
known_for=pd.read_sql("""
  SELECT *
  FROM known_for
  """, conn)
known_for.info()
```

```python
movie_akas=pd.read_sql("""
  SELECT *
  FROM movie_akas
  """, conn)
movie_akas.info()
```

```python
movie_ratings=pd.read_sql("""
  SELECT *
  FROM movie_ratings
  """, conn)
movie_ratings.info()
```

```python
persons=pd.read_sql("""
  SELECT *
  FROM movie_basics
  """, conn)
persons.info()
```

```python
principals=pd.read_sql("""
  SELECT *
  FROM principals
  """, conn)
principals.info()
```

```python
writers=pd.read_sql("""
  SELECT *
  FROM writers
  """, conn)
writers.info()
```
