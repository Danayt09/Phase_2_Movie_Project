```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
from zipfile import ZipFile # for unzipping the db.zip file
pd.set_option('display.max_columns', None) # show all columns when printing out dataframe
```

```python
import os
import zipfile

# Check if 'im.db' exists, unzip if it doesn't
if not os.path.exists('zippedData/im.db'):
    with zipfile.ZipFile('zippedData/im.db.zip', 'r') as zip_ref:
        zip_ref.extractall('zippedData')

# Connect to the SQLite database
conn = sqlite3.connect('zippedData/im.db')
```

```python
persons= pd.read_sql("""
  SELECT *
  FROM persons
  """, conn)
persons
```

```python
# Dropping 'birth_year' and 'death_year' columns
persons.drop(['birth_year', 'death_year'], axis=1, inplace=True)
```

```python
print(persons.columns)
```

```python
# Replace existing 'persons' table with updated DataFrame
persons.to_sql('persons', conn, if_exists='replace', index=False)
```
